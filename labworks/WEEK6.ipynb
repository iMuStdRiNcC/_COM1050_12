```python
#W6A10 
num_list = list(map(int,input().split()))
k = int(input())

list_of_sum = []

for i in num_list:
    for n in num_list:
        sum = i + n 
         
        if sum == k:
            tuple_pair = (i,n)
            num_list.append(tuple_pair)
        else:
            continue

print(list_of_sum)
```

```python
#W6A11
num_list = list(map(int,input().split()))
even = []
odd = []
for number in num_list:
    if number % 2 ==0:
        even.append(number)
    else :
        odd.append(number)

print(tuple(even))
print(tuple(odd))
```

```python
#W6A12
num_list = list(input().split())
num_dict = {}
for num in num_list:
    if num in num_dict:
        new_occ = num_dict[num] + 1 
        num_dict[num] = new_occ
    else:
        num_dict.update({num:1})

count = 0

for occ in list(num_dict.values()):
    if count == 0:
        most_occ = occ
        count += 1
    else:
        if occ > most_occ:
            most_occ = occ
            count += 1
        else:
            count += 1
            pass

occ_list = list(num_dict.values())
index = occ_list.index(most_occ)
key_list = list(num_dict.keys())
print(key_list[index])
```

```python
#W6A13
seq = list(input().split())
random_dict = {}
for i in range(len(seq))[::2]:
    if i == len(seq)-1:
        random_dict.update({"":seq[i]})
    else:
        random_dict.update({seq[i+1]:seq[i]})

print(random_dict)
```

```python
#W6A15
seq = list(input().split())
random_dict = {}
for i in range(len(seq))[::2]:
    if i == len(seq)-1:
        continue
    else:
        if int(seq[i+1]) >90:
            random_dict.update({seq[i]:seq[i+1]})
        else:
            continue 

print(random_dict)
```

```python
#W6A21
seq = list(map(int,input().split()))
dict = {}
for i in seq:
    dict.update({i:[]})
    for k in range(len(seq)):
        if i == seq[k]:
            dict[i].append(k)
        else:
            continue 

for char in seq:
    dict[char] = tuple(dict[char])

print(dict)
```
